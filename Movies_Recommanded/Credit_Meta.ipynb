```python
import pandas as pd
import numpy as np
```

```python
credit_df = pd.read_csv('credits.csv')
```

```python
credit_df.head()
```

```python
meta_df = pd.read_csv('movies_metadata.csv')
```

```python
meta_df.head(3)
```

```python
meta_df['release_date'] = pd.to_datetime(meta_df['release_date'], errors='coerce')
```

```python
meta_df['year'] = meta_df['release_date'].dt.year
```

```python
meta_df['year'].value_counts().sort_index()
```

```python
new_meta = meta_df.loc[meta_df['year'] == 2017, ['genres','id','title','year']]
```

```python
new_meta.head()
```

```python
new_meta.id = new_meta.id.astype(int)
```

```python
df = pd.merge(new_meta, credit_df, on = 'id')
```

```python
df.head()
```

```python
pd.set_option('display.max_colwidth', 80)
df.head()
```

```python
import ast
df['genres'] = df['genres'].map(lambda x: ast.literal_eval(x))
df['cast'] = df['cast'].map(lambda x: ast.literal_eval(x))
df['crew'] = df['crew'].map(lambda x: ast.literal_eval(x))
```

```python
def make_genresList(x):
    gen = []
    st = " "
    for i in x:
        if i.get('name') == 'Science Fiction':
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    if gen == []:
        return np.NaN
    else:
        return (st.join(gen))
```

```python
df['genres_list'] = df['genres'].map(lambda x: make_genresList(x))
```

```python
df['genres_list']
```

```python
def get_actor1(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == []:
        return np.NaN
    else:
        return (casts[0])
df['actor_1_name'] = df['cast'].map(lambda x: get_actor1(x))
```

```python
def get_actor2(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=1:
        return np.NaN
    else:
        return (casts[1])
df['actor_2_name'] = df['cast'].map(lambda x: get_actor2(x))
```

```python
def get_actor3(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=2:
        return np.NaN
    else:
        return (casts[2])
df['actor_3_name'] = df['cast'].map(lambda x: get_actor3(x))
```

```python
def get_directors(x):
    dt = []
    st = " "
    for i in x:
        if i.get('job') == 'Director':
            dt.append(i.get('name'))
    if dt == []:
        return np.NaN
    else:
        return (st.join(dt))
df['director_name'] = df['crew'].map(lambda x: get_directors(x))
```

```python
df.columns
```

```python
movie = df.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres_list','title']]
movie.head()
```

```python
movie.isnull().sum()
```

```python
movie = movie.dropna(how = 'any')
```

```python
movie.isnull().sum()
```

```python
movie = movie.rename(columns={'genres_list':'genres'})
movie['title'] = movie['title'].str.lower()
movie['combination'] = movie['actor_1_name'] + ' ' + movie['actor_2_name'] + ' '+ movie['actor_3_name'] + ' '+ movie['director_name'] +' ' + movie['genres']
```
